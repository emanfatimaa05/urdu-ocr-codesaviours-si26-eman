## Question 1

### What is OCR (Optical Character Recognition)?

OCR (Optical Character Recognition) is a technology that converts printed or handwritten text in images into editable digital text. It allows computers to recognize characters from scanned documents, books, newspapers, and photographs. OCR helps save time by reducing manual typing and making printed information searchable and editable.

## Question 2

### Why is Urdu OCR harder than English OCR?

Urdu OCR is more difficult than English OCR because Urdu is written from right to left and many letters change their shape depending on their position in a word. Many Urdu characters also look very similar and contain small dots that can easily be missed by an OCR system. Different handwriting styles, fonts, and calligraphy make recognizing Urdu text even more challenging.

## Question 3

### What are two real-world applications of Urdu OCR?

Urdu OCR can be used to digitize old Urdu books, newspapers, and historical documents so they can be searched and preserved digitally. It can also help convert printed government forms, educational material, and business documents into editable text, making information easier to access and process.

```python
from PIL import Image, ImageDraw, ImageFont
import os

urdu_texts = [
    "پاکستان زندہ باد",
    "آج کا موسم خوشگوار ہے",
    "تعلیم ہر انسان کا حق ہے",
    "کراچی پاکستان کا سب سے بڑا شہر ہے",
    "محنت کامیابی کی کنجی ہے",
    "پانی زندگی کے لیے ضروری ہے",
    "کتاب علم کا خزانہ ہے",
    "وقت کی قدر کرنی چاہیے",
    "سچ بولنا ایک اچھی عادت ہے",
    "لاہور تاریخی شہر ہے",
    "صحت سب سے بڑی دولت ہے",
    "دوستی ایک انمول رشتہ ہے",
    "درخت لگانا ماحول کے لیے اچھا ہے",
    "اردو ہماری قومی زبان ہے",
    "محبت سے دل جیتے جاتے ہیں",
    "استاد قوم کا معمار ہوتا ہے",
    "صبر کا پھل میٹھا ہوتا ہے",
    "بچوں کی تربیت اہم ذمہ داری ہے",
    "خوشی بانٹنے سے بڑھتی ہے",
    "امن ہر معاشرے کی ضرورت ہے"
]

os.makedirs('data/raw/synthetic', exist_ok=True)
font = ImageFont.truetype('/content/NotoNastaliqUrdu.ttf', 28, layout_engine=ImageFont.Layout.RAQM)

for i, text in enumerate(urdu_texts):
    img = Image.new('RGB', (400, 80), color='white')
    draw = ImageDraw.Draw(img)
    draw.text((10, 20), text, fill='black', font=font, direction='rtl')
    img.save(f'data/raw/synthetic/urdu_{i+1}.png')
    print(f'Generated: urdu_{i+1}.png -- {text}')

print(f'\nDone! {len(urdu_texts)} images saved in data/raw/synthetic/ folder')
```
