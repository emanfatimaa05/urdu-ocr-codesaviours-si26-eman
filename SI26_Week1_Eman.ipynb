### What is OCR (Optical Character Recognition)?

OCR (Optical Character Recognition) is a technology that converts printed or handwritten text in images into editable digital text. It allows computers to recognize characters from scanned documents, books, newspapers, and photographs. ## Question 2

### Why is Urdu OCR harder than English OCR?

Urdu OCR is more difficult than English OCR because Urdu is written from right to left and many letters change their shape depending on their position in a word. Many Urdu characters also look very similar and contain small dots that can easily be missed by an OCR system. Different handwriting styles, fonts, and calligraphy make recognizing Urdu text even more challenging.

### What are two real-world applications of Urdu OCR?

Urdu OCR can be used to digitize old Urdu books, newspapers, and historical documents so they can be searched and preserved digitally. It can also help convert printed government forms, educational material, and business documents into editable text, making information easier to access and process.
